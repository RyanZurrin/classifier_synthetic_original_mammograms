```python
import numpy as np
import matplotlib.pyplot as plt
import os

# Specify the directory containing your NPZ files
npz_dir = '/content/drive/MyDrive/original'

# Specify the directory where you want to save the PNG files
png_dir = '/content/drive/MyDrive/OG'

# Create the output directory if it doesn't exist
os.makedirs(png_dir, exist_ok=True)

# List all NPZ files in the directory
npz_files = [f for f in os.listdir(npz_dir) if f.endswith('.npz')]

for npz_file in npz_files:
    try:
        # Load data from the NPZ file
        data = np.load(os.path.join(npz_dir, npz_file))

        # Get a list of keys in the NPZ file
        keys = data.files
        print(f"Keys in {npz_file}: {keys}")

        # Loop through the keys and display their data (if they are images)
        for key in keys:
            image_data = data[key]
            # You can add logic to check if the data is an image and save it accordingly
            if isinstance(image_data, np.ndarray) and len(image_data.shape) == 2:
                # Create a PNG image from the data
                plt.imshow(image_data, cmap='gray')
                plt.axis('off')
                plt.savefig(os.path.join(png_dir, f'{npz_file}_{key}.png'), bbox_inches='tight', pad_inches=0)
                plt.close()
    except Exception as e:
        print(f"Error processing {npz_file}: {e}")



```
